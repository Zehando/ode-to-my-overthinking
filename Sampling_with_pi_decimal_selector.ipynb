```python
import pandas as pd
import numpy as np
from mpmath import mp

url = 'https://drive.google.com/file/d/1bvqIxFdYVcKIqb8HvmFHzpq-1Vt7PqQ2/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
```

```python
def add_unique_random_numbers(df, sample_number, pi_digits, start_range=1):
  num_row = len(df)
  end_range = start_range + num_row
  num_digit = len(str(end_range))
  all_numbers = np.arange(start_range, end_range)
  random_numbers = np.random.choice(all_numbers, size=num_row, replace=False)
  random_numbers_formatted = [str(num).zfill(num_digit) for num in random_numbers]
  df['random'] = random_numbers_formatted
  num_digit = len(df.iloc[1,1])
  pi_list = []
  mp.dps = pi_digits
  pi_num = mp.pi
  pi_str = str(pi_num)
  decimals = pi_str.split('.')[1]
  for i in range(0, len(decimals), len(df.iloc[1,1])):
    parts = decimals[i:i+(len(df.iloc[1,1]))]
    pi_list.append(parts)
  match_count = 0
  df['random_check'] = False
  for pi_digit_string in pi_list:
    if match_count < sample_number:
      matching_indices = df[(df['random'] == pi_digit_string) & (df['random_check'] == False)].index


      if len(matching_indices) > 0:
        first_match_index = matching_indices[0]
        df.loc[first_match_index, 'random_check'] = True
        match_count += 1
    else:
      break

  return df.loc[df['random_check']==True]
```

```python
df_with_random = add_unique_random_numbers(df, 30, 10000)
df_with_random
```
